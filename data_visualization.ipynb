## Data Visualization on Heart Disease Data

Importing libraries:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Read in data:

```python
heart = pd.read_csv('heart_disease_uci.csv')
```

### Comparing number of major vessels observed to the stage of heart disease (ca vs. num)

```python
no_disease = heart[heart['num'] == 0]
disease_one = heart[heart['num'] == 1]
disease_two = heart[heart['num'] == 2]
disease_three = heart[heart['num'] == 3]
disease_four = heart[heart['num'] == 4]

fig, ax = plt.subplots(1, 5, figsize=(21, 6))

ax1 = sns.countplot(data=no_disease, x='ca', ax=ax[0])
ax1.set(xlabel='Number of Major Vessels', ylabel='Count', title='No Heart Disease')

ax2 = sns.countplot(data=disease_one, x='ca', ax=ax[1])
ax2.set(xlabel='Number of Major Vessels', ylabel='Count', title='Stage One Heart Disease')

ax3 = sns.countplot(data=disease_two, x='ca', ax=ax[2])
ax3.set(xlabel='Number of Major Vessels', ylabel='Count', title='Stage Two Heart Disease')

ax4 = sns.countplot(data=disease_three, x='ca', ax=ax[3])
ax4.set(xlabel='Number of Major Vessels', ylabel='Count', title='Stage Three Heart Disease')

ax5 = sns.countplot(data=disease_four, x='ca', ax=ax[4])
ax5.set(xlabel='Number of Major Vessels', ylabel='Count', title='Stage Four Heart Disease')
```

### Comparing chest pain (cp) vs. num

```python
fig, ax = plt.subplots(2, 3, figsize=(35, 13))
sns.set_palette('muted')

ax1 = sns.countplot(data=no_disease, y='cp', ax=ax[0, 0])
ax1.set(xlabel='Count', ylabel='Chest Pain', title='No Heart Disease')

ax2 = sns.countplot(data=disease_one, y='cp', ax=ax[0, 1])
ax2.set(xlabel='Count', ylabel='Chest Pain', title='Stage One Heart Disease')

ax3 = sns.countplot(data=disease_two, y='cp', ax=ax[0, 2])
ax3.set(xlabel='Count', ylabel='Chest Pain', title='Stage Two Heart Disease')

ax4 = sns.countplot(data=disease_three, y='cp', ax=ax[1, 0])
ax4.set(xlabel='Count', ylabel='Chest Pain', title='Stage Three Heart Disease')

ax5 = sns.countplot(data=disease_four, y='cp', ax=ax[1, 1])
ax5.set(xlabel='Count', ylabel='Chest Pain', title='Stage Four Heart Disease')

fig.delaxes(ax[1][2])
```

```python
g = sns.countplot(data=heart, x='sex', hue='cp')
g.set_yscale("log")
g.set(xlabel='Gender', ylabel='Log of Count', title='Gender vs. Count of Types of Chest Pain')
g.legend(title='Chest Pain Types')
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15, 11))

typical_angina_heart = heart[heart['cp'] == 'typical angina']
ax0 = sns.countplot(data=typical_angina_heart, x='num', hue='sex', ax=ax[0, 0])
ax0.set(xlabel='Stage of Heart Disease', ylabel='Count',
      title='Typical Angina Chest Pain vs. Heart Disease Progression by Gender')
ax0.legend(title='Gender', loc='upper right')

atypical_angina_heart = heart[heart['cp'] == 'atypical angina']
ax1 = sns.countplot(data=atypical_angina_heart, x='num', hue='sex', ax=ax[0, 1])
ax1.set(xlabel='Stage of Heart Disease', ylabel='Count',
      title='Atypical Angina Chest Pain vs. Heart Disease Progression by Gender')
ax1.legend(title='Gender', loc='upper right')

non_anginal_heart = heart[heart['cp'] == 'non-anginal']
ax2 = sns.countplot(data=non_anginal_heart, x='num', hue='sex', ax=ax[1, 0])
ax2.set(xlabel='Stage of Heart Disease', ylabel='Count',
      title='Non-Anginal Chest Pain vs. Heart Disease Progression by Gender')
ax2.legend(title='Gender', loc='upper right')

asymptomatic_heart = heart[heart['cp'] == 'asymptomatic']
ax3 = sns.countplot(data=asymptomatic_heart, x='num', hue='sex', ax=ax[1, 1])
ax3.set(xlabel='Stage of Heart Disease', ylabel='Count',
      title='Asymptomatic Chest Pain vs. Heart Disease Progression by Gender')
ax3.legend(title='Gender', loc='upper right')
```

### Comparing ST depression (oldpeak) vs. num

```python
ax0 = sns.catplot(data=heart, x='num', y='oldpeak', kind='box')
ax0.set(xlabel='Stage of Heart Disease', ylabel='ST Depression Induced by Exercise, Relative to Rest',
      title='ST Depression Throughout Stages of Heart Disease')

ax1 = sns.catplot(data=heart, x='num', y='oldpeak', kind='violin')
ax1.set(xlabel='Stage of Heart Disease', ylabel='ST Depression Induced by Exercise, Relative to Rest',
      title='ST Depression Throughout Stages of Heart Disease')
```

### Age vs. num

```python
heart_disease = heart[heart['num'] != 0]
disease_one = heart[heart['num'] == 1]
disease_two = heart[heart['num'] == 2]
disease_three = heart[heart['num'] == 3]
disease_four = heart[heart['num'] == 4]


g = sns.histplot(data=heart_disease, x='age', hue='num', kde=True)
g.set(xlabel='Age', title='Age vs. Progression of Heart Disease')
g.legend(title='Heart Disease Stage', labels=['1', '2', '3', '4'])
```

```python
disease_one['age'].mean(axis=0), disease_two['age'].mean(axis=0), disease_three['age'].mean(axis=0), disease_four['age'].mean(axis=0)
```

```python
under55 = heart[heart.age < 55]
above55 = heart[heart.age >= 55]

fig, ax = plt.subplots(1, 2, figsize=(13, 6))

ax0 = sns.countplot(data=above55, x='num', hue='ca', log=True, ax=ax[1])
ax0.legend(title='Number of Major Vessels', loc='upper right')
ax0.set(xlabel='Stage of Heart Disease', ylabel='Log of Count', title='Number of Major Vessels vs. Stage of Heart Disease Above 55')

ax1 = sns.countplot(data=under55, x='num', hue='ca', log=True, ax=ax[0])
ax1.legend(title='Number of Major Vessels', loc='upper right')
ax1.set(xlabel='Stage of Heart Disease', ylabel='Log of Count', title='Number of Major Vessels vs. Stage of Heart Disease Under 55')
```

```python
typical_angina_heart = heart[heart['cp'] == 'typical angina']
non_anginal_heart = heart[heart['cp'] == 'non-anginal']

fig, ax = plt.subplots(1, 2, figsize=(13, 6))

ax1 = sns.histplot(data=typical_angina_heart, x='age', ax=ax[0], kde=True)
ax1.set(xlabel='Age', ylabel='Count', title='Typical Anginal Chest Pain over Age')
ax1.set_xticks(range(30, 80, 5))

ax2 = sns.histplot(data=non_anginal_heart, x='age', ax=ax[1], kde=True)
ax2.set(xlabel='Age', ylabel='Count', title='Non-Anginal Chest Pain over Age')
ax2.set_xticks(range(30, 80, 5))
```
